```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv("nonlinear_data.csv")
X = dataset.iloc[ : , 0:-1 ].values
y = dataset.iloc[ : , -1 ].values
```

## Training the simple linear regression model on the training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

## Predicting the linear regression results

```python
predicted_y= regressor.predict(X)
```

## Visualizing the linear regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Linear regression ')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Creating polynomial features of the independent variable

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 6)
X_poly = poly_reg.fit_transform(X)
```

```python
print(X_poly)
```

## Training the Polynomial Regression Model on the Whole Dataset

```python
regressor_poly= LinearRegression()
regressor_poly.fit(X_poly, y)
```

## Predicting the polynomial regression results

```python
predicted_y_poly = regressor_poly.predict(X_poly)
```

```python
print(y)
```

```python
print(predicted_y_poly)
```

## Visualizing the polynomial regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor_poly.predict(X_poly), color = 'blue')
plt.title('Polynomial regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
