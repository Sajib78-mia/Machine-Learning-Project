# Salary Prediction using Linear Regression

```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Salary Data.csv", sep=',')
df
```

```python
plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Company Salary')
```

```python
x=df[['Experience Years']]
x
```

```python
y=df[['Salary']]
y
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Company Salary')
```

```python
reg.fit(x,y) ##train the model
```

```python
m=reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
m*16.7+c
```

```python
reg.predict([[16.7]])
```

```python
df['predicted']=reg.predict(x)
df
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Company Salary')
```

# Cost & Lose

```python
df['Lose']=df['Salary']-df['predicted']
df
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mse=mean_squared_error(df['Salary'],df['predicted'])
mse
```

```python
mae=mean_absolute_error(df['Salary'],df['predicted'])
mae
```

```python
reg.score(x,y)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,reg.predict(x))
```
