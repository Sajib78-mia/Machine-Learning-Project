```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
House_price_dataset= pd.read_csv("House_price_data.csv")
House_price_dataset.head()
```

```python
X= House_price_dataset.iloc[:,2:10].values
y= House_price_dataset.iloc[:,1].values
y = [int(label) for label in y] 
```

```python
from pandas.core.common import random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier= SVC(kernel='linear',random_state = 0)
classifier.fit(X_train,y_train) 
```

```python
## accauracy on training set
X_train_prediction= classifier.predict(X_train)
X_train_prediction
training_accuracy_score= accuracy_score(y_train,X_train_prediction)
print(training_accuracy_score)
```
